```python
'''
Podemos fazer testes a respeito de faixas de valores criadas por nós mesmos.
Lidando com o contexto de criação de trilhas para aprendizados, é importante saber do nível de 
experiência dos diferentes desenvolvedores, sendo assim, vou filtrar a quantidade de profissionais de 
acordo com os anos de exeperiência, visando os diferentes tipos de hierarquia adotados globalmente:

0-2 anos: júnior
2-5 anos: pleno
5-8 anos: sênior
8+: tech lead
10+: arquiteto/gestor
'''

'''
Primeiramente, para filtrar esses dados, precisamos tratar os casos especiais, que, na nossa coluna
correspondente aos anos de experiência, são dados mal preenchidos com "mais de tantos anos" ou "NA"
'''
#Resolvendo essa questão, temos:
# Primeiro, substitua os valores não numéricos por números equivalentes
base_devs['YearsCode'] = base_devs['YearsCode'].replace({
    'Less than 1 year': '0',
    'More than 50 years': '51'
})

# Agora converte para número
base_devs['YearsCode'] = pd.to_numeric(base_devs['YearsCode'], errors='coerce')

#E, finalmente, podemos montar as desigualdades:

contagem_junior = (base_devs['YearsCode'] <=2).sum() #Aqui, fazemos uma contagem de todos os valores
                                                     #menores que 2 anos de experiência
contagem_junior
```

```python
contagem_pleno = ((base_devs['YearsCode'] > 2) & 
            (base_devs['YearsCode'] <= 5)).sum() #Aqui, fazemos o mesmo, mas entre dois valores
contagem_pleno
```

```python
contagem_senior = ((base_devs['YearsCode'] > 5) & 
            (base_devs['YearsCode'] <= 8)).sum() #Repetimos
contagem_senior
```

```python
contagem_tech_lead = ((base_devs['YearsCode'] > 8) & 
            (base_devs['YearsCode'] <= 10)).sum() #Mais uma vez
contagem_senior
```

```python
contagem_junior = (base_devs['YearsCode'] >10).sum() #Por fim, os mais experientes
contagem_junior
```
